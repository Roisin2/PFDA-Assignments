```python
import numpy as np
import matplotlib.pyplot as plt

# Function to mimic a single battle round
def mimic_battle_round():
    attacker_dice = np.sort(np.random.randint(1, 7, size=3))[-2:]
    defender_dice = np.sort(np.random.randint(1, 7, size=2))
    attacker_losses = np.sum(attacker_dice <= defender_dice)
    defender_losses = 2 - attacker_losses
    return attacker_losses, defender_losses

# mimic 1000 battle rounds
rounds = 1000
attacker_losses = []
defender_losses = []
for _ in range(rounds):
    a_loss, d_loss = mimic_battle_round()
    attacker_losses.append(a_loss)
    defender_losses.append(d_loss)

# Plot the results
plt.hist(attacker_losses, bins=np.arange(0, 4) - 0.5, alpha=0.7, label='Attacker Losses', color='red')
plt.hist(defender_losses, bins=np.arange(0, 3) - 0.5, alpha=0.7, label='Defender Losses', color='blue')
plt.xticks(range(4))
plt.xlabel('Losses per Round')
plt.ylabel('Frequency')
plt.title('Risk Battle Mock-Up (3 Attackers vs 2 Defenders)')
plt.show
```
